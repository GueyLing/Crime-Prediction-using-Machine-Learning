```python
def binary_encoding(col_name):
    # Perform Binary Numerical Encoding
    category_to_binary = {}
    binary_repr = []

    for category in balanced_data[col_name]:
        if category not in category_to_binary:
            category_to_binary[category] = bin(len(category_to_binary))[2:].zfill(len(bin(len(category_to_binary))) - 2)
        binary_repr.append(category_to_binary[category])

    # Store binary representation in a single feature
    balanced_data[col_name] = binary_repr
    balanced_data[col_name] = balanced_data[col_name].astype(int)
```

```python
from itertools import product
param_grid = {
    'iterations': [1000, 1500],
    'learning_rate': [0.05, 0.1, 0.2],
    'depth': [6, 8, 10]
}

# Create a list of all hyperparameter combinations
param_combinations = list(product(*param_grid.values()))

# Initialize variables to track the best result
best_params = None
best_test_accuracy = -float('inf')  # Change this to the appropriate classification metric

# Iterate through each hyperparameter combination
for params in param_combinations:
    hyperparameters = dict(zip(param_grid.keys(), params))

    # Create a CatBoostClassifier with the current hyperparameters
    catboost_model = CatBoostClassifier(**hyperparameters, random_state=42, verbose=False)

    # Train the model on the training data
    catboost_model.fit(X_train_scaled, y_train)

    # Predict on the test set
    y_pred_test = catboost_model.predict(X_test_scaled)

    # Calculate the accuracy on the test set (you can change this to another classification metric)
    test_accuracy = accuracy_score(y_test, y_pred_test)

    # Print the hyperparameters and the corresponding accuracy (or other classification metric)
    print("Hyperparameters:", hyperparameters)
    print("Test Accuracy:", test_accuracy)
    print("-" * 50)

    # Check if this is the best result so far
    if test_accuracy > best_test_accuracy:
        best_test_accuracy = test_accuracy
        best_params = hyperparameters

# Print the best hyperparameters and corresponding best test accuracy (or other classification metric)
print("Best Hyperparameters:", best_params)
print("Best Test Accuracy:", best_test_accuracy)
```
